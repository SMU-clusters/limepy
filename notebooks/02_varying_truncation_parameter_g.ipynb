## Varying truncation parameter g

Compare density profiles for models with different truncation parameters *g*. Some well known models are included: g=0 (Woolley 1956); g=1 (King 1966);  g=2 (Wilson 1975).

```python
%matplotlib inline
import matplotlib.pyplot as plt
from limepy import limepy
import numpy
plt.loglog()
plt.axis([1e-2, 1e3, 1e-8, 1e1])
g = numpy.linspace(0,3,7)
for i in range(len(g)):
    m = limepy(5,g[i])
    plt.plot(m.r, m.rho)
```
