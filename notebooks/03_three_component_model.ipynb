## Three component model

Here we construct a simple 3 component model (model A from [Gunn & Griffin 1979](http://adsabs.harvard.edu/abs/1979AJ.....84..752G)). Generating the multi-mass model is almost as straightforward as the single-mass model. We just need to define a mass function 

```python
from limepy import limepy
mj = [0.5, 1.0, 1.5]
Mj = [5, 1, 0.1]
m = limepy(12, 1, mj=mj, Mj=Mj, M=5e5, rh=12,project=True)
```

(Note: if no scales for M and r are set, all densities are normalised to the central value.) Plotting the projected densities (as in Fig. 3a of GG79) 

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.loglog()
plt.axis([0.1, 1e3, 1e-6, 1e1])
for j in range(3):
    plt.plot(m.r/m.r0, m.Sigmaj[j]/m.Sigma[0])
```

The line-of-sight velocity dispersions of the 3 components are

```python
import numpy
for j in range(3):
    plt.plot(m.r/m.r0, numpy.sqrt(m.v2pj[j]))
```
